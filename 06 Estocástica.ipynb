## Distribución uniforme

```python
import numpy as np

# Número de muestras
N = 1000000

# Sea X una variable aleatoria uniforme 0,1
x = np.random.rand(N)

print(x)
x.mean() # O bien: x.sum() / N
```

```python
import matplotlib.pyplot as plt

plt.hist(x, bins=10, range=[0,1])
plt.title('Histograma')
```

## Distribución normal

```python
N = 1000000

# Distribución normal con media 0 y desviación est. de 1
x = np.random.randn(N)

plt.hist(x, bins=1000, density=True)
plt.title('Histograma')
```

```python
# Distribución normal con media mu y desviación est. de sigma
mu = 4
sigma = 3

x = mu + sigma * np.random.randn(N)

plt.hist(x, bins=1000, density=True)
plt.title('Histograma')
```

Función de densidad: 
$$f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

```python
x = np.arange(mu-4*sigma, mu+4*sigma,.1)
y = 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-((x-mu)**2/(2*sigma**2)))

plt.plot(x,y)
plt.title("Función de densidad de la dist. normal")
```

## Estimación de $\pi$

El número $\pi$ se puede estimar de diferentes maneras. En esta clase veremos dos:
- Haciendo uso del cálculo de la integral numérica por medio de rectángulos
- Usando el método de Montecarlo

### Integral numérica

Existen métodos más sofisticados para calcular la integral numéricamente, por ejemplo con el método del trapecio. Para ello se puede usar la función [`trapz`](https://numpy.org/doc/1.25/reference/generated/numpy.trapz.html).

En este caso, sin embargo, lo haremos tomando la integral de Riemann a través de los rectángulos formados por la función.

Conocemos que la función de la circunferencia unitaria es:
$$x^2+y^2=1$$

Por tanto, podremos tomar la función:
$$f(x)=\sqrt{x^2-1}$$
e integrarla:
$$\int_0^1 f(x)dx = \frac{\pi}{4}$$

Por lo que nuestro $\pi$ estimado será:

$$\pi = 4\int_0^1 \sqrt{x^2-1} dx$$

```python
N = 10000
x = np.arange(0,1,1/N)

y = np.sqrt(1-x**2)

plt.plot(x,y)
```

```python
# Estimación de pi usando la suma de Riemann
pi_est = y.sum()/N*4

print(f"El número pi estimado es: {pi_est}, con un error de {np.abs(pi_est-np.pi)}")
```

## Método de Montecarlo

```python
N = 1_000_000

x = np.random.rand(N)
y = np.random.rand(N)

dentro = (x**2+y**2<1)

# Número de puntos dentro
num_dentro = dentro.sum()

print(f"Cayeron {num_dentro} puntos dentro del cuarto de círculo de los {N}")

pi_est = num_dentro / N * 4
print(f"El número pi estimado es: {pi_est}, con un error de {np.abs(pi_est-np.pi)}")
```
