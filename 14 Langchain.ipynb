```python
from langchain.agents import create_agent

def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always rainy in {city}!"

def get_latest_tech_info(question: str) -> str:
    """Get information for technological news."""
    return f"The latest iPhone is iPhone 25. {question}"

agent = create_agent(
    model="gpt-5-mini",
    tools=[get_weather, get_latest_tech_info],
    system_prompt="You are a helpful assistant. Always use one of the given tools.",
)

# Run the agent
res = agent.invoke(
    {"messages": [{"role": "user", "content": "Cual es el tiempo en Madrid?"}]}
)

res
```
